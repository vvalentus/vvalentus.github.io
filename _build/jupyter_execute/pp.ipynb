```python
import pandas as pd
import warnings
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import mean
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('athlete_events.csv')
df.head()
```

Tal como se indicó en las instrucciones, se filtró el dataframe para tomar aquellos deportistas que participaron en el año 2016 y que ganaron una medalla.

```python
df2016m =  df.query('Year == 2016')
df2016m =  df2016m[df2016m['Medal'].notna()]
df2016m.head()
```

Para obtener un recuento de las medallas por deporte en el 2016 se hizo uso de value_counts(), método que arroja por defecto los resultados en orden descendente.

```python
df2016m['Sport'].value_counts()
```

Donde los 5 deportes con más medallas fueron *atletismo*, *natación*, *remo*, *fútbol* y *hockey*. Ahora filtremos el dataframe en función de los 5 deportes con más medallas, quedándonos solo con cinco deportes.

```python
sports = ['Athletics', 'Swimming', 'Rowing', 'Football', 'Hockey']
df2016m = df2016m[df2016m['Sport'].isin(sports)]
df2016m.head()
```

Al graficar el subconjunto de datos filtrado, se obtienen los mismos resultados y en el mismo orden arrojados por value_counts().

```python
sns.set_theme(style="whitegrid")

ax = sns.countplot(x="Sport", data=df2016m, palette="Set2", order=sports)
ax.set(xlabel="Deporte", ylabel="Medallas olímpicas", 
       title="5 deportes con más medallas olímpicas en 2016")
ax.set_xticklabels(["Atletismo", "Natación", "Remo", "Fútbol", "Hockey"], fontsize=10)

plt.show()
```

También se realizó la gráfica para visualizar la distribución de la edad entre los medallistas de estos deportes en 2016. 

```python
ax = sns.histplot(df2016m['Age'], palette = "Set2")
ax.set(title = 'Distribución de la edad de medallistas olímpicos en 2016', xlabel = "Edad", ylabel = 'Frecuencia')
```

Se puede observar que hay un gran número de medallistas entre los 22 y 33 años aproximadamente. De igual forma, la cantidad de medallistas menores de 19 años y mayores de 33 es mucho menor. 

Ahora revisemos cómo están distruidas el número de medallas por países, haciendo uso nuevamente de value_counts()

```python
df2016m['Team'].value_counts()
```

```python
orden = df2016m['Team'].value_counts().index

plt.figure(figsize=(14, 6))
ax = sns.countplot(x="Team", data=df2016m, palette="Set2", order=orden)
ax.set(xlabel="Equipo", ylabel="Medallas olímpicas", 
       title="Medallas olímpicas por país")
plt.xticks(rotation=90)
plt.show()
```

Tal como lo indica el recuento de países ganadores de medallas, Estados Unidos es el país con más medallas (127 medallas), seguido por Alemania (88 medallas) y Gran Bretaña (69 medallas) de 54 países en total. 

```python
print(mean(df2016m[df2016m['Sex']== 'F']['Weight']))
print(mean(df2016m[df2016m['Sex']== 'M']['Weight']))

print(df2016m[df2016m['Sex']== 'F'].size)
print(df2016m[df2016m['Sex']== 'M'].size)
```

También se realizó un análisis del peso promedio por sexo, se puede notar una gran diferencia entre los pesos de medallistas mujeres (5325 medallistas) con un peso promedio de 65.24kg y medallistas hombres (5655) con un peso promedio de 81.8kg.

```python
ax = sns.barplot(x = "Sex", y = "Weight", data = df2016m, estimator = mean, palette = 'Set2');
ax.set(xlabel = 'Sexo', ylabel = 'Peso', title = 'Peso promedio por sexo')
```

Vemos que esa diferencia de precio se mantiene independientemente del deporte. Aún así, el peso promedio de ganadores de medallas olímpicas en 2016 en remo parecen ser más pesados en ambos sexos comparados con otros deportes. En deportes como fútbol, atletismo y hockey los pesos son muy similares.

```python
plt.figure(figsize=(10, 6))
ax = sns.barplot(x = "Sport", y = "Weight", hue = "Sex", data = df2016m, estimator = mean, palette = 'Set2', order= sports);
ax.set(xlabel = 'Sexo', ylabel = 'Peso', title = 'Peso promedio por sexo y deporte')
ax.set_xticklabels(["Atletismo", "Natación", "Remo", "Fútbol", "Hockey"], fontsize=10)
```
